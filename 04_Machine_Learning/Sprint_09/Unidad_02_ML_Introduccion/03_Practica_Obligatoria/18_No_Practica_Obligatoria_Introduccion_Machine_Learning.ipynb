## Cientfico de datos, qu hace?

## Flujo de un proyecto de ML
1. **Problema de negocio**: entender bien el problema  intentamos transformarlo en un problema tcnico (traducir la necesidad del negocio a un problema abordable con datos)  

2. **Definir el tipo de problema (en este caso es una prediccin)**
- **Regresin**: variable objetivo numrica
- **Clasificacin**: variable objetivo categrica
- **No supervisado**: *clustering*, generacin de textos, anlisis de textos, etc.  

3. **Acceso a los datos**
Fuentes posibles: BBDD, APIs, web scraping, CSV, etc.  

4. **Limpieza y transformacin de datos**
- Tratamiento de valores nulos y outliers **solo con datos de TRAIN** para evitar el data leakage (fuga de datos)
- Identificacin del **target** (ej. `median_house_value`)  

5. **Separacin de datos: TRAIN / TEST (/ VALIDATION (especfico de Deep Learning))**
- El conjunto TEST simula datos que el modelo "no conoce"
- Divisin **estratificada** cuando sea necesario (especialmente en clasificacin, ej. dividida en 3 categoras)  

6. **EDA (Mini-EDA) sobre TRAIN**
- Anlisis del target (relacin del target con el resto de variables)
- Relacin del target  

- **Seleccin y creacin de features**
- Comprensin del funcionamiento del modelo y **explicabilidad** (a la hora de seleccionar el modelo con el que nos quedamos, el que consuma menos recursos, el que sea ms explicable, ser el mejor. Ej. regresin lineal vs deep learning)  

7. **Tratamiento de features**
- Variables categricas  numricas
- Variables continuas  escalado (si el mtodo lo requiere)  

8. **Seleccin de modelos y mtricas (Baseline)**
- Definir modelos candidatos
- Elegir mtricas:
    - Clasificacin  Accuracy, Precision, Recall, F1 (til en desbalanceo), Balanced Accuracy
    - Regresin  MAE, MSE, RMSE, MAPE, R  

9. **Entrenamiento del modelo**
- Ajuste inicial: `fit`  

10. **Ajuste de hiperparmetros**
- Bsqueda del mejor conjunto de parmetros (Grid Search, Random Search, etc.)  

11. **Regularizacin**
- Reducir el **overfitting** (hemos adecuado tanto los datos del train, que el modelo reacciona muy mal ante los del test)
- Objetivo principal: **generalizacin** (equilibrio entre bias y variance) (sweet spot en amarillo)  

  
12. **Evaluacin final en TEST**
- Mtrica en TEST (*Test Error*)
- Comparacin **Mtrain vs Mtest**
- Diagnstico: si no generaliza  **overfitting**
