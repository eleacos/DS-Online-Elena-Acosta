### Clase vs Objeto
[al indice](#Contenidos)  

**La clase se usa para definir algo**. Al igual que con las funciones. Creamos el esqueleto de lo que ser un objeto de esa clase. Por tanto, **una vez tengamos la clase definida, instanciaremos un objeto de esa clase**.  Es como crear el concepto de coche, con todas sus caractersticas y funcionalidades. Despus, a lo largo del programa, podremos crear objetos de tipo coche, que se ajusten a lo definido en la clase coche. Cada coche tendr una marca, una potencia, etc

````python
primer_coche = Coche()
print(type(primer_coche))
print(primer_coche)
````

Ahora s tenemos un objeto de tipo Coche, que se llama `primer_coche`. Cuando imprimimos su tipo, vemos que es de tipo Coche, y cuando lo imprimes el objeto por pantalla, simplemente nos dice su tipo y un identificador.

Podremos crear todos los coches que queramos

````python
citroen = Coche()
peugeot = Coche()

print(citroen == peugeot)
````
